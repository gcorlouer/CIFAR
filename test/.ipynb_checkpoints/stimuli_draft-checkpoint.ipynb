```python
from mne_bids import make_bids_basename
from pathlib import Path, PurePath
import seaborn
import helper
import mne
import os
import matplotlib.pyplot as plt
import pandas as pd
import matlab.engine
import numpy as np
import scipy as sp
import scipy.io
import re
pd.options.display.html.table_schema = True
pd.options.display.max_rows = None
from mne.time_frequency import psd_multitaper
%matplotlib 

# TODO 
```

```python
proc = 'BP'
subid = 'DiAs'
cfsubdir = Path('~','CIFAR_data','iEEG_10','subjects', subid, 'EEGLAB_datasets', 'bipolar_montage').expanduser()
bids_subpath = Path('~','projects','CIFAR','data_bids', 'sub-04' ).expanduser()
elecinfo = bids_subpath.joinpath('anat','electrodes_info.csv')
dfelec = pd.read_csv(elecinfo)
ROIs = dfelec['Brodman'].unique()
nROIs = len(ROIs)
fname = helper.CIFAR_filename(subid=subid,task='stimuli',proc=proc)
fpath = cfsubdir.joinpath(fname)
fpath = os.fspath(fpath)
raw = mne.io.read_raw_eeglab(fpath, preload=True)
```

```python
raw.plot(duration=30, n_channels=20, scalings=2e-4, color='b', butterfly=False)
```

```python
# High frequency band extraction
# ts parameters
nobs = len(raw[0][1])
nchan = len(raw.info['ch_names'])
# gamma bands
l_freq = 60
nband = 6
band_size = 20
bands = [l_freq+i*band_size for i in range(1, nband)]
#print(bands)

HFB_norm = np.zeros(shape=(nobs, nchan))
HFB_mean_amplitude = np.zeros(shape=(nchan,))

for band in bands:
    raw_band = raw.copy().filter(l_freq=band, h_freq=band+band_size, phase='minimum', filter_length='auto', l_trans_bandwidth= 10, h_trans_bandwidth= 10, 
                                 fir_window='blackman')
    HFB = raw_band.copy().apply_hilbert(envelope=True).to_data_frame().to_numpy()
    HFB_mean_amplitude = HFB_mean_amplitude + np.mean(HFB, axis=0)
    HFB = np.divide(HFB, np.mean(HFB, axis=0))
    HFB_norm = HFB_norm + HFB 

HFB_norm = HFB_norm/nband
HFB_mean_amplitude = HFB_mean_amplitude/nband
HFB_norm = HFB_norm * HFB_mean_amplitude
HFB_norm = np.nan_to_num(HFB_norm)
```

```python
# Create raw_HFB structure 
HFB_norm = np.transpose(HFB_norm) # Confert to good format
raw_HFB = mne.io.RawArray(HFB_norm, raw.info)
```

```python
# Create annotations from initial raw
raw_HFB.set_annotations(raw.annotations)
events, events_id = mne.events_from_annotations(raw_HFB)
```

```python
# Return faces and places id
p_face = re.compile('Face')
p_place = re.compile('Place')
place_id = []
face_id = []
#p_face.match(events_id.keys())
for key in events_id.keys():
    if p_place.match(key):
        place_id.append(str(events_id[key]))
    if p_face.match(key):
        face_id.append(str(events_id[key]))
```

```python
# Face and (?) place selective electrodes
picks = ['LTo1-LTo2', 'LTo5-LTo6']
```

```python
epochs = mne.Epochs(raw_HFB, events, tmin=-0.5, tmax=2, 
                    event_repeated='drop', preload=True)
```

```python
# Plot psd
epochs[face_id[0]].plot_psd(xscale='log', picks='LTo1-LTo2', color='blue')
```

```python
# Plot histogram
V2_elec = epochs[face_id[0]].get_data(picks='LTo1-LTo2')
data = V2_elec[0,0,:]
hist = np.histogram(data, bins=100, density=True)
V2_elec.shape
plt.hist(data, bins=100)
plt.show()
plt.title('Histogram of HFB envelope response of visualy responsive electrode')
```

```python
data = V2_elec[0,0,:]
log_data = np.log(np.absolute(data))
plt.hist(log_data, bins=100)
plt.show()
```

```python
# Fit log normal distrib
from scipy.stats import lognorm
x = np.linspace(-1.5,15, num=1000)
data = V2_elec[0,0,:]
mu = np.mean(V2_elec[0,0,:])
s = np.std(V2_elec[0,0,:])
param = scipy.stats.lognorm.fit(data)
pdf = scipy.stats.lognorm.pdf(x, param[0], loc=param[1], scale=param[2])
#plt.hist(V2_elec[0,0,:], bins=100)
plt.plot(pdf)
plt.plot(hist[0])
```

```python
evok_face = epochs[face_id].average()
evok_place = epochs[place_id].average()
```

```python
# Get face selective and place selective ERP
# Face and (?) place selective electrodes (LTp in fusiform)
picks = ['LTo1-LTo2', 'LTp3-LTp4','LFa5-LFa6', 'LTo5-LTo6']
epochs_picks = mne.Epochs(raw_HFB, events, tmin=-0.5, tmax=2, picks=picks,  
                    event_repeated='drop', preload=True)
```

```python
# Save epoched data
save2 = Path('~','projects','CIFAR','data_fun').expanduser()
fname_face = helper.CIFAR_filename(subid=subid,task='stimuli',proc=proc, suffix='_epoch_face.mat')
fname_place = helper.CIFAR_filename(subid=subid,task='stimuli',proc=proc, suffix='_epoch_place.mat')
fpath_face = save2.joinpath(fname_face)
fpath_place = save2.joinpath(fname_place)
epochs_picks_face = epochs_picks[face_id].get_data()
epochs_picks_place = epochs_picks[place_id].get_data()
scipy.io.savemat(fpath_face, dict(epochs_picks_face=epochs_picks_face))
scipy.io.savemat(fpath_place, dict(epochs_picks_place=epochs_picks_place))
```

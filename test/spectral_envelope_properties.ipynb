```python
raw_rest, dfelec = helper.import_data(task='rest_baseline')
raw_stim, _ = helper.import_data(task='stimuli')
```

```python
raw_band = raw_rest.copy().filter(l_freq=65, h_freq=85, phase='minimum', filter_length='auto', l_trans_bandwidth= 10, h_trans_bandwidth= 10, 
                                     method='iir', iir_params=dict(order=70, ftype='butter', output='sos'))
envelope_band = raw_band.copy().apply_hilbert(envelope=True)
```

```python
envelope_band.plot_psd(picks=picks)
```

```python
picks = ['LGRD49-LGRD57', 'LGRD50-LGRD51', 'LGRD58-LGRD59','LTo1-LTo2','LTo3-LTo4','LTo5-LTo6','LTp1-LTp2','LTp3-LTp4']
raw_band = raw_band.get_data(picks=picks)
envelope_band = envelope_band.get_data(picks=picks)
```

```python
psd_raw, freqs = psd_multitaper(raw_band, picks=picks) 
psd_envelope_band,freqs = psd_multitaper(envelope_band, picks=picks)

psd_raw = 20.*np.log10(psd_raw)
psd_envelope_band = 20.*np.log10(psd_envelope_band)
psd_raw_mean = psd_raw.mean(0)
psd_envelope_mean = psd_envelope_band.mean(0)
plt.plot(freqs, psd_raw_mean, label='psd filtered LFP')
plt.plot(freqs, psd_envelope_mean, label='psd envelope')
plt.xscale('linear')
plt.xlim((0.1,200))
plt.xlabel('freqs (Hz)')
plt.ylabel('Amplitude (dB)')
plt.title('Power spectral density of bipolar LFP and envelope comparison in 65-85 Hz band')
plt.legend()
```

```python
HFB_rest, raw_HFB_rest = helper.HFB(raw_rest, l_freq=60, nband=6, band_size=20);
HFB_stim, raw_HFB_stim = helper.HFB(raw_stim, l_freq=60, nband=6, band_size=20);
```

```python
# Interesting electrodes
picks = ['LGRD49-LGRD57', 'LGRD50-LGRD51', 'LGRD58-LGRD59','LTo1-LTo2','LTo3-LTo4','LTo5-LTo6','LTp1-LTp2','LTp3-LTp4']
raw_HFB_stim = mne.io.RawArray(HFB_stim, raw_stim.info)
raw_HFB_rest = mne.io.RawArray(HFB_rest, raw_stim.info)

# Create annotations from initial raw
raw_HFB_stim.set_annotations(raw_stim.annotations)
events, events_id = mne.events_from_annotations(raw_HFB_stim)
```

```python
# Return faces and places id
p_face = re.compile('Face')
p_place = re.compile('Place')
place_id = []
face_id = []
#p_face.match(events_id.keys())
for key in events_id.keys():
    if p_place.match(key):
        place_id.append(str(events_id[key]))
    if p_face.match(key):
        face_id.append(str(events_id[key]))
```

```python
epochs_rest_picks = mne.Epochs(raw_HFB_rest, events, tmin=-0.5, tmax=2, picks=picks,  
                    event_repeated='drop', preload=True);
epochs_stim_picks = mne.Epochs(raw_HFB_stim, events, tmin=-0.5, tmax=2, picks=picks,  
                    event_repeated='drop', preload=True);
```

```python
epochs_LFP_picks = mne.Epochs(raw_rest, events, tmin=-0.5, tmax=2, picks=picks,   
                    event_repeated='drop', preload=True)
```

```python
# Compare psd
plt.figure()
epochs_rest_picks.plot_psd(xscale='log',show=False)
epochs_LFP_picks.plot_psd(xscale='log', show=False)
```

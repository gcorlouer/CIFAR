```python
raw_rest, dfelec = helper.import_data(task='rest_baseline')
raw_stim, _ = helper.import_data(task='stimuli')
```

```python
dfelec.loc[dfelec['Brodman'] == 'MT']
```

```python
HFB_rest, raw_HFB_rest = helper.HFB(raw_rest, l_freq=60, nband=6, band_size=20)
HFB_stim, raw_HFB_stim = helper.HFB(raw_stim, l_freq=60, nband=6, band_size=20)
```

```python
# Interesting electrodes
picks = ['LGRD60-LGRD61', 'LGRD49-LGRD57', 'LGRD50-LGRD51', 'LGRD58-LGRD59','LTo1-LTo2','LTo3-LTo4','LTo5-LTo6','LTp1-LTp2','LTp3-LTp4']
raw_HFB_stim = mne.io.RawArray(HFB_stim, raw_stim.info)
raw_HFB_rest = mne.io.RawArray(HFB_rest, raw_stim.info)

# Create annotations from initial raw
raw_HFB_stim.set_annotations(raw_stim.annotations)
events, events_id = mne.events_from_annotations(raw_HFB_stim)
```

```python
raw_HFB_stim.plot(duration=30, n_channels=20, scalings=1, color='b', butterfly=False)
```

```python
# Return faces and places id
p_face = re.compile('Face')
p_place = re.compile('Place')
place_id = []
face_id = []
#p_face.match(events_id.keys())
for key in events_id.keys():
    if p_place.match(key):
        place_id.append(str(events_id[key]))
    if p_face.match(key):
        face_id.append(str(events_id[key]))
```

```python
epochs_rest_picks = mne.Epochs(raw_HFB_rest, events, tmin=-0.1, tmax=2.5, picks=picks,  
                    event_repeated='drop', preload=True)
epochs_stim_picks = mne.Epochs(raw_HFB_stim, events, tmin=-0.1, tmax=2.5, picks=picks,  
                    event_repeated='drop', preload=True)
```

```python
time = epochs_rest_picks.times
```

```python
evok_stim_face = epochs_stim_picks[face_id].average()
evok_rest_face = epochs_rest_picks[face_id].average()
evok_rest_face.plot(picks='LGRD60-LGRD61')
```

```python
# Select place and face selective epochs

epoch_stim_face = epochs_stim_picks[face_id]
epoch_stin_place = epochs_stim_picks[place_id]
```

```python
baseline_stim =  epochs_stim_picks[face_id].crop(tmin=-0.4, tmax=-0.1)
baseline_stim = baseline_stim.average()
```

```python
ERP_stim = evok_stim_face.data
ERP_baseline = baseline_stim.data
ERP_rest = evok_rest_face.data
```

```python
baseline_shift_rest = 10*np.log10(1+np.divide(np.square(ERP_stim[5,:]),np.mean(np.square(ERP_baseline[5,:]))))
```

```python
plt.rcParams.update({'font.size': 17})
```

```python
# Plot baseline shift

plt.plot(time, baseline_shift_rest)
plt.xlabel('Time locked at stimulus onset (s)')
plt.ylabel('Amplitude (dB)')
plt.title('Baseline shift of high frequency amplitude ERP for representative face selective electrodes in V2 from stimulus onset')
```
